```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_reviews=pd.read_csv("googleplaystore_user_reviews.csv")
raw_reviews.head()
```

```python
raw_reviews.shape
```

```python
#Checking missing values
raw_reviews.isnull().sum()
```

```python
raw_reviews=raw_reviews.dropna(subset=["Translated_Review"],axis=0).reset_index().drop(["index"],axis=1)
```

```python
raw_reviews.isnull().sum()
```

```python
raw_reviews.shape
```

```python
raw_reviews.head()
```

```python
raw_reviews.dtypes
```

```python
#check that string values are in str type
raw_reviews[["App","Translated_Review","Sentiment"]]=raw_reviews[["App","Translated_Review","Sentiment"]].astype("string")
```

```python
#rounding Sentiment_Subjectivity for better readibility
raw_reviews["Sentiment_Subjectivity"]=round(raw_reviews["Sentiment_Subjectivity"],3)
```

```python
raw_reviews.dtypes
```

```python
raw_reviews.head()
```

```python
#Loading the cleaned dataset
cleaned_reviews=raw_reviews.copy()
cleaned_reviews.to_csv("Cleaned_reviews.csv",index=False)
```
